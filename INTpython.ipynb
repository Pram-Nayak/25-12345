```python
import nltk
import random
from bs4 import BeautifulSoup
from collections import defaultdict

# Read the data
with open('/content/drive/MyDrive/Colab Notebooks2/NLP/data/electronics/positive.review', 'r', encoding='utf-8') as file:
    positive_reviews = BeautifulSoup(file.read(), "html.parser")

pos_reviews = positive_reviews.findAll('review_text')
```

```python
print(pos_reviews[1:10])
```

```python
from builtins import range
import nltk
nltk.download('punkt_tab')
```

```python
#Here we extract the trigrams from positive reviews dataset and we intert them into the dictionary
# Extract trigrams and store in dictionary
grams = defaultdict(list)

for review in pos_reviews:
    s = review.text.lower()
    tokens = nltk.word_tokenize(s)
    for i in range(len(tokens) - 2):
        k = (tokens[i], tokens[i + 2])  #trigram i + 2
        grams[k].append(tokens[i + 1])
```

```python
grams
```

```python
# Convert trigrams into probability distributions
# probability calculation
# The frequency information is converted into probabilities.
probabilities = {}

for k, words in grams.items():
    freq_dist = nltk.FreqDist(words)
    probabilities[k] = {word: freq_dist.freq(word) for word in freq_dist}
```

```python
probabilities
```

```python
# Function to generate a spun review
#here a random review is selected by using trigrams there is a 20 percent chance that it'll be modified
def test_spinner():
    review = random.choice(pos_reviews)
    s = review.text.lower()
    print("Original Text:", s)

    tokens = nltk.word_tokenize(s)
    for i in range(len(tokens) - 2):
        if random.random() < 0.2:  # 20% chance of replacement
            k = (tokens[i], tokens[i + 2])
            if k in probabilities:
                tokens[i + 1] = random.choice(list(probabilities[k].keys()))

    spun_text = " ".join(tokens)
    spun_text = spun_text.replace(" .", ".").replace(" ,", ",").replace(" '", "'").replace("$ ", "$").replace(" !", "!")

    print("\nSpin Text:")
    print(spun_text)
```

```python
test_spinner()
```

## Article spinner for both positive and negative reviews

```python
from google.colab import drive
import nltk
import random
from bs4 import BeautifulSoup
from collections import defaultdict
from builtins import range
```

```python

# Read the data
with open('/content/drive/MyDrive/Colab Notebooks2/NLP/data/electronics/positive.review', 'r', encoding='utf-8') as file:
    positive_reviews = BeautifulSoup(file.read(), "html.parser")

with open('/content/drive/MyDrive/Colab Notebooks2/NLP/data/electronics/negative.review', 'r', encoding='utf-8') as file:
    negative_reviews = BeautifulSoup(file.read(), "html.parser")

pos_reviews = positive_reviews.findAll('review_text')
neg_reviews = negative_reviews.findAll('review_text')
```

```python
print(pos_reviews[1:10])
```

```python
print(neg_reviews[1:10])
```

```python
nltk.download('punkt_tab')
```

```python
# Function to extract trigrams and probabilities from reviews
def extract_trigram_probabilities(reviews):
  grams = defaultdict(list)
  for review in reviews:
      s = review.text.lower()
      tokens = nltk.word_tokenize(s)
      for i in range(len(tokens) - 2):
          k = (tokens[i], tokens[i + 2])
          grams[k].append(tokens[i + 1])

  probabilities = {}
  for k, words in grams.items():
      freq_dist = nltk.FreqDist(words)
      probabilities[k] = {word: freq_dist.freq(word) for word in freq_dist}
  return probabilities

# Extract trigrams and probabilities for positive and negative reviews
pos_probabilities = extract_trigram_probabilities(pos_reviews)
neg_probabilities = extract_trigram_probabilities(neg_reviews)
```

```python
# Function to generate a spun review
def test_spinner(reviews, probabilities):
    review = random.choice(reviews)
    s = review.text.lower()
    print("Original Text:", s)

    tokens = nltk.word_tokenize(s)
    for i in range(len(tokens) - 2):
        if random.random() < 0.2:  # 20% chance of replacement
            k = (tokens[i], tokens[i + 2])
            if k in probabilities:
                tokens[i + 1] = random.choice(list(probabilities[k].keys()))

    spun_text = " ".join(tokens)
    spun_text = spun_text.replace(" .", ".").replace(" ,", ",").replace(" '", "'").replace("$ ", "$").replace(" !", "!")

    print("\nSpin Text:")
    print(spun_text)

```

```python
print("\n--- Positive Review Spinner ---")
test_spinner(pos_reviews, pos_probabilities)
```

```python
print("\n--- Negative Review Spinner ---")
test_spinner(neg_reviews, neg_probabilities)
```
